**Text Cleaning and Vectorization with CountVectorizer:**

This code defines a function `clean_and_vectorize` that performs text cleaning and vectorization using the `CountVectorizer` from scikit-learn. The function takes a dataset (`data`) and a text column (`text_column`), cleans the text by removing stopwords and punctuation, and then applies the CountVectorizer with a custom analyzer to tokenize and vectorize the text. The output includes the feature names extracted from the vectorizer and the transformed matrix of text features. This process is essential for preparing text data for machine learning tasks such as classification or clustering.


````python
# Define the main function
def clean_and_vectorize(data, text_column):
  from sklearn.feature_extraction.text import CountVectorizer
  from nltk.corpus import stopwords
  def text_cleaning(text):

    stop_words = set(stopwords.words('english'))
    punctuationfree = "".join([i for i in text if i not in string.punctuation])
    stopwordfree = [word for word in punctuationfree.split() if word.lower() not in stop_words]
    return stopwordfree

  # Create an instance of CountVectorizer with the custom analyzer
  vectorizer = CountVectorizer(analyzer=text_cleaning)

  # Fit and transform the data
  reviews_count = vectorizer.fit_transform(data[text_column])

  # Get the feature names and the transformed matrix
  feature_names = vectorizer.get_feature_names_out()
  transformed_matrix = reviews_count.toarray()

  return feature_names, transformed_matrix
````

````python
# Use the function
feature_names, transformed_matrix = clean_and_vectorize(data, 'verified_reviews')
````

This function, `convert_boolean_to_binary`, iterates through each column in a DataFrame (`df`). If a column's data type is boolean (`bool`), it converts the boolean values to binary integers (`0` for `False` and `1` for `True`) using `astype(int)`. The modified DataFrame, `reviews_df`, is returned with boolean columns converted to binary format. This transformation is useful for standardizing boolean data representations, ensuring compatibility with various data analysis and machine learning algorithms.



````python
def convert_boolean_to_binary(df):
  for column in df:
    if df[column].dtype == bool:
      df[column] = df[column].astype(int)
  return df

reviews_df = convert_boolean_to_binary(reviews_df)
reviews_df.head()
````

After importing these libraries, you can proceed with further steps like splitting data into training and testing sets, fitting the model, predicting with the test set

````python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
````



**Fit Logistic Regression Model**

This code snippet initializes a logistic regression model (`model`) using `LogisticRegression` from `sklearn.linear_model` and fits it to the training data (`X_train` and `y_train`). Logistic regression is a linear model that is widely used for binary classification tasks.



````python
model = LogisticRegression()
model.fit(X_train, y_train)
````

This code snippet predicts the labels for the test data (`X_test`) using the trained logistic regression model (`model`).
````python
model.predict(X_test)
````
